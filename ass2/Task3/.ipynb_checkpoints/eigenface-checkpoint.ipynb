```python
import numpy as np
import cv2
import os
import matplotlib.pyplot as plt
```

# 2.1

```python
path='./Yale-FaceA/trainingset/'
trainList = os.listdir(path)
FFF = True
for i in trainList:
    tmp = path+i
#     print(i)
    img = plt.imread(tmp)
#     print(img.shape)
    img = img.reshape(-1)
    if FFF:
        Train_data = img.copy()
        FFF = False
        continue
    Train_data = np.row_stack((Train_data,img))
Train_data = Train_data.T
Train_data.shape
```

# 2.2

```python
pix,num=Train_data.shape
meanface = Train_data[:,0].copy()
now = Train_data.copy()
Qnorms = Train_data[:,0].copy()

for i in range(pix):
    meanface[i] = now[i].sum()/num
    now[i] = now[i] - meanface[i]
    tmp = abs(now[i])
    if max(tmp)!=0:
        Qnorms[i]=max(tmp)
    else:
        Qnorms[i]=1
    now[i]=now[i]/Qnorms[i]
plt.title('meanface')
plt.imshow(meanface.reshape((231,195)),cmap='gray')

co = np.cov(now.T@now)
D,vec=np.linalg.eig(co)
eigenface = now@vec[:,:15]
eigenface=np.array(eigenface)
eigenface=np.real(eigenface)
eigenface.shape
```

# 2.3

```python
plt.figure(figsize=(30, 12))
for i in range(15):
    plt.subplot(3,5,i+1)
    plt.imshow(eigenface[:,i].reshape(231,195),cmap='gray')
# plt.savefig('eigenface14')
plt.show()
```

# 2.4

```python
for i in range(eigenface.shape[1]):
    eigenface[:,i]=eigenface[:,i]/np.sqrt(np.sum(np.square(eigenface[:,i])))
corr = now.T@eigenface

path='./Yale-FaceA/testset/'
trainList = os.listdir(path)
FFF = True
for file in trainList:
    tmp = path+file
    refer = plt.imread(tmp)
    refer = refer.reshape(-1)
    for i in range(pix):
        refer[i]-=meanface[i]
        refer[i]/=Qnorms[i]
    img=(refer.T@eigenface@eigenface.T*Qnorms+meanface).reshape(-1)
#     plt.imshow(img.reshape(231,195),cmap='gray')

    target = [ np.sum(( ((refer-meanface)/Qnorms).T@eigenface - ((now[:,i]-meanface)/Qnorms).T@eigenface )**2) for i in range(num) ]
    tar_copy = target.copy()
    tar_copy.sort()

    plt.figure(figsize=(20, 6))
    plt.subplot(1,4,1)
    plt.imshow(refer.reshape(231,195),cmap='gray')
    for i in range(3):
        plt.subplot(1,4,i+2)
        plt.imshow(Train_data[:, target.index( tar_copy[i]) ].reshape(231,195),cmap='gray')
```
